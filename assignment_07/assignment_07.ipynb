```python
import pandas as pd
import plotly.express as px
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/wcj365/python-stats-dataviz/master/assignments/assignment_07_world_dev/world_devlopment_indicator.csv")
df.shape
```

```python
df.sample(5)
```

```python
df.tail(5)
```

```python
df["Country Name"].unique()
```

```python
df2=df[df['Country Name'].isin(['Australia','India','New Zealand','United States'])]
```

```python
df2.shape
```

```python
df2.sample(5)
```

```python
fig = px.bar(
    df2.query('indicator == "NY.GDP.PCAP.PP.CD" and Year == 2021'), 
    x="Country Name", 
    y="value", 
    title="2021 GDP of Australia, India, New Zealand and USA", 
    color="Country Name", 
    template="plotly_dark"
)

fig.update_layout(showlegend=False)

fig.show()
```

```python
fig = px.pie(
    df2.query('indicator == "NY.GDP.PCAP.PP.CD" and Year == 2021'),  
    names="Country Name", 
    values="value", 
    title="2021 GDP of Australia, India, New Zealand and USA", 
    color="Country Name", 
    template="plotly_dark"
)

fig.update_layout(showlegend=True)

fig.update_traces(textposition='inside', textinfo='label+percent+value')

fig.show()
```

```python
fig = px.line(
    df2.query('indicator == "NY.GDP.PCAP.PP.CD"'),
    x="Year",
    y="value", 
    title="GDP of Australia, India, New Zealand and USA over time", 
    color="Country Name", 
    template="plotly_dark"
)

fig.update_layout(showlegend=True)

fig.show()
```

```python
df.groupby(['Region','indicator']).mean()
#df.groupby(["state", "gender"], as_index=False)["last_name"].count()
```

```python
my_indicator=["NY.GDP.PCAP.PP.CD"]
df5 = df[df["indicator"].isin(my_indicator)]
```

```python
df5
```

```python
df2021 = df5.query("Year==2021")
df2021
```

```python
df_region_2021=df2021.groupby(['Region']).mean()
df_region_2021.reset_index(inplace=True)
```

```python
fig = px.bar(
    df_region_2021, 
    x="Region", 
    y="value", 
    title="2021 GDP of regions", 
    color="Region", 
    template="plotly_dark"
)

fig.update_layout(showlegend=False)

fig.show()
```

```python
fig = px.pie(
    df_region_2021, 
    names="Region", 
    values="value", 
    title="2021 GDP of Regions", 
    color="Region", 
    template="plotly_dark"
)

fig.update_layout(showlegend=True)

fig.update_traces(textposition='inside', textinfo='label+percent+value')

fig.show()
```

```python
df_region=df5.groupby(['Region','Year']).mean()
df_region.reset_index(inplace=True)
```

```python
fig = px.line(
    df_region,
    x="Year",
    y="value", 
    title="GDP of region", 
    color="Region", 
    template="plotly_dark"
)

fig.update_layout(showlegend=True)

fig.show()
```

```python
my_indicators = ["SP.DYN.LE00.IN", "NY.GDP.PCAP.PP.CD"]

df6 = df[df["indicator"].isin(my_indicators)]

df6.shape
```

```python
df5.sample(10)
```

```python
df6 = df6[["Year",	"value",	"indicator",	"Country Code",	"Country Name"]]
          
df6.sample(5)
```

```python
df7 = pd.pivot_table(df6, values = 'value', index=['Country Code', 'Country Name','Year'], columns = 'indicator').reset_index()

df7.head()
```

```python
df7.dropna(inplace=True)

df7.head()
```

```python
fig = px.scatter(
    df7.query('Year == 2020'),
    x="NY.GDP.PCAP.PP.CD",
    y="SP.DYN.LE00.IN", 
    title="Wealth vs Health", 
    color="Country Name", 
    template="plotly_dark"
)

fig.update_layout(showlegend=False)

fig.show()
```
